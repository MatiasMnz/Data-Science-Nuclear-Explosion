# Introducción

Las explosiones nucleares han sido un tema de gran importancia en la historia moderna, tanto por su impacto en la geopolítica mundial como por sus consecuencias en términos de seguridad global y medio ambiente. El estudio y análisis de las explosiones nucleares no solo arroja luz sobre los eventos pasados, sino que también nos brinda información valiosa para comprender las implicaciones presentes y futuras de la energía nuclear y las decisiones políticas relacionadas. En este contexto, el análisis de datos se presenta como una herramienta fundamental para desentrañar patrones, tendencias y relaciones ocultas dentro de los registros históricos de explosiones nucleares.

El propósito de este trabajo es explorar y analizar detalladamente las explosiones nucleares registradas a lo largo del tiempo, centrándonos en tres objetivos principales:

## 1. Análisis de Patrones Temporales y Geográficos:
   Investigaremos y visualizaremos la distribución temporal y geográfica de las explosiones nucleares registradas. Nuestro objetivo es identificar patrones estacionales, tendencias a lo largo del tiempo y áreas geográficas más afectadas por estas explosiones. Al comprender mejor cómo y dónde han ocurrido estas explosiones, podremos arrojar luz sobre los factores subyacentes que han influido en su distribución.

## 2. Predicción de Potencia y Efectos:
   Desarrollaremos modelos predictivos para estimar la potencia de las explosiones nucleares y predecir sus posibles efectos. Utilizaremos variables como la magnitud, ubicación y condiciones ambientales para crear modelos precisos que puedan proporcionar información sobre la intensidad de las explosiones y sus posibles consecuencias. Esta predicción podría contribuir a una mejor comprensión de los riesgos asociados con las explosiones nucleares.

## 3. Evaluación de Causas y Consecuencias:
   Realizaremos un análisis causal para comprender las causas subyacentes de las explosiones nucleares y sus posibles consecuencias. Examinaremos cómo factores políticos, sociales y tecnológicos han influido en la ocurrencia y prevención de estas explosiones. Al identificar posibles relaciones causales, podremos contribuir al debate sobre la proliferación nuclear y la seguridad internacional.

Mediante el análisis detallado de datos, buscamos arrojar luz sobre estos aspectos cruciales de las explosiones nucleares, contribuyendo así a una comprensión más profunda de su impacto y su papel en la historia y la seguridad global.

## Objetivo 1: Análisis de Patrones Temporales y Geográficos:

```python
# Pregunta 1: Estacionalidad en la frecuencia de explosiones nucleares a lo largo de los años
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='Date.Year')
plt.title('Estacionalidad en la Frecuencia de Explosiones Nucleares a lo largo de los Años')
plt.xlabel('Año')
plt.ylabel('Cantidad de Explosiones')
plt.xticks(rotation=45)
plt.show()
```

#### Breve Analisis
El gráfico muestra la cantidad de explosiones nucleares registradas en cada año. Se observa que la frecuencia de explosiones nucleares ha variado a lo largo de los años. Algunos años presentan picos de explosiones nucleares, mientras que en otros años la cantidad es menor.

```python
# Insight 1: Estacionalidad en la Frecuencia de Explosiones Nucleares a lo largo de los Años
print("Insight 1: Estacionalidad en la Frecuencia de Explosiones Nucleares a lo largo de los Años")
year_with_max_explosions = data['Date.Year'].value_counts().idxmax()
max_explosions = data['Date.Year'].value_counts().max()
year_with_min_explosions = data['Date.Year'].value_counts().idxmin()
min_explosions = data['Date.Year'].value_counts().min()
print(f"Año con mayor número de explosiones: {year_with_max_explosions} con {max_explosions} explosiones.")
print(f"Año con menor número de explosiones: {year_with_min_explosions} con {min_explosions} explosiones.")
```

```python
# Pregunta2: ¿Cómo ha evolucionado la magnitud de las explosiones nucleares a lo largo de los años?
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='Date.Year', y='Data.Magnitude.Body', marker='o')
plt.title('Evolución de la Magnitud de Explosiones Nucleares a lo largo de los Años')
plt.xlabel('Año')
plt.ylabel('Magnitud del Evento (Kilotones)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

#### Breve Analisis
El gráfico muestra cómo la magnitud de explosiones nucleares aumentó en las décadas de 1950 y 1960, posiblemente debido a avances tecnológicos y tensiones políticas. A partir de la década de 1970, disminuyó, quizás debido a tratados de limitación de armas. Esto refleja la interacción compleja entre desarrollos técnicos y acuerdos diplomáticos en la historia nuclear.

```python
# Insight: Evolución de la Magnitud de las Explosiones Nucleares
max_magnitude_year = data.groupby('Date.Year')['Data.Magnitude.Body'].max().idxmax()
max_magnitude_value = data.groupby('Date.Year')['Data.Magnitude.Body'].max().max()

min_magnitude_year = data.groupby('Date.Year')['Data.Magnitude.Body'].min().idxmin()
min_magnitude_value = data.groupby('Date.Year')['Data.Magnitude.Body'].min().min()

print("Insight: Evolución de la Magnitud de Explosiones Nucleares")
print(f"Año con la mayor magnitud de explosión: {max_magnitude_year} con {max_magnitude_value} kilotones.")
print(f"Año con la menor magnitud de explosión: {min_magnitude_year} con {min_magnitude_value} kilotones.")
```

```python
# Pregunta 3: Identificación de áreas geográficas con mayor número de explosiones nucleares
plt.figure(figsize=(12, 6))
scatter = sns.scatterplot(data=data, x='Location.Cordinates.Longitude', y='Location.Cordinates.Latitude', hue='WEAPON DEPLOYMENT LOCATION', palette='Set1', alpha=0.7)
plt.title('Identificación de Áreas Geográficas con Mayor Número de Explosiones Nucleares')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
scatter.legend(title='País de Despliegue', loc='upper right', bbox_to_anchor=(1.3, 1))
plt.show()
```

#### Breve Analisis
En este gráfico de dispersión, cada punto representa una explosión nuclear en función de su latitud y longitud. Los puntos están coloreados según el país de despliegue. Se puede observar que algunas áreas geográficas presentan una concentración significativamente mayor de explosiones nucleares, lo que podría indicar regiones más afectadas.

```python
# Insight 3: Identificación de Áreas Geográficas con Mayor Número de Explosiones Nucleares
print("Insight 3: Identificación de Áreas Geográficas con Mayor Número de Explosiones Nucleares")
area_with_max_explosions = data['WEAPON DEPLOYMENT LOCATION'].value_counts().idxmax()
max_area_explosions = data['WEAPON DEPLOYMENT LOCATION'].value_counts().max()
area_with_min_explosions = data['WEAPON DEPLOYMENT LOCATION'].value_counts().idxmin()
min_area_explosions = data['WEAPON DEPLOYMENT LOCATION'].value_counts().min()
print(f"Área con mayor número de explosiones: {area_with_max_explosions} con {max_area_explosions} explosiones.")
print(f"Área con menor número de explosiones: {area_with_min_explosions} con {min_area_explosions} explosiones.")
```

```python
# Pregunta 1: ¿Se puede desarrollar un modelo preciso para predecir la potencia de una explosión nuclear en función de diversas variables?
print("Pregunta 1: Modelo de Predicción de Potencia")
avg_potency = data['Data.Yeild.Upper'].mean()
max_potency = data['Data.Yeild.Upper'].max()
min_potency = data['Data.Yeild.Upper'].min()

plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='Data.Yeild.Upper', bins=20, kde=True)
plt.title('Distribución de Potencia de Explosiones Nucleares')
plt.xlabel('Potencia (kilotones)')
plt.ylabel('Frecuencia')
plt.show()
```

#### Breve Análisis
En el gráfico de distribución de potencia, podemos observar que la mayoría de las explosiones nucleares tienen una potencia relativamente baja, con unos pocos casos de potencias excepcionalmente altas

```python
# Insight
print("Insight 1: Modelo de Predicción de Potencia")
print(f"Potencia promedio de explosiones nucleares: {avg_potency:.2f} kilotones")
print(f"Potencia máxima registrada: {max_potency:.2f} kilotones")
print(f"Potencia mínima registrada: {min_potency:.2f} kilotones")
```

```python
# Pregunta 2: ¿Cuál es la influencia de la magnitud y las condiciones ambientales en la potencia de las explosiones nucleares?
print("\nPregunta 2: Influencia de Magnitud en Potencia")
correlation_magnitude_potency = data['Data.Magnitude.Body'].corr(data['Data.Yeild.Upper'])

plt.figure(figsize=(10, 6))
sns.jointplot(data=data, x='Data.Magnitude.Body', y='Data.Yeild.Upper', kind='hex', cmap='viridis')
plt.suptitle('Influencia de la Magnitud en la Potencia de Explosiones Nucleares', y=1.02)
plt.xlabel('Magnitud de la Explosión (Cuerpo)')
plt.ylabel('Potencia (kilotones)')
plt.show()
```

#### Breve Análisis
En el gráfico de hexágonos con paleta de colores 'viridis', podemos observar la distribución de densidad de puntos entre la magnitud de la explosión en el cuerpo y la potencia de la explosión.

El gráfico de hexágonos representa la relación entre la magnitud de la explosión en el cuerpo y la potencia de las explosiones nucleares. Los hexágonos más oscuros indican áreas con una mayor concentración de explosiones. Observamos que existe una tendencia general de aumento en la potencia a medida que la magnitud en el cuerpo aumenta. Las barras en la parte superior y derecha detallan las distribuciones individuales de ambas variables.

```python
# Insight
print("Insight 2: Influencia de Magnitud en Potencia")
print(f"Correlación entre Magnitud y Potencia: {correlation_magnitude_potency:.2f}")
```

```python
# Pregunta 3: ¿Podemos estimar con precisión los efectos de una explosión nuclear en función de su potencia y ubicación?
print("\nPregunta 3: Estimación de Efectos según Potencia y Ubicación")
avg_surface_magnitude = data['Data.Magnitude.Surface'].mean()
max_surface_magnitude = data['Data.Magnitude.Surface'].max()
min_surface_magnitude = data['Data.Magnitude.Surface'].min()

# Visualización
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(data=data, x='Data.Yeild.Upper', y='Data.Magnitude.Surface', hue='WEAPON DEPLOYMENT LOCATION')
plt.title('Estimación de Efectos según Potencia y Magnitud en la Superficie')
plt.xlabel('Potencia (kilotones)')
plt.ylabel('Magnitud en la Superficie')
plt.legend(title='País de Despliegue')
scatter_plot.legend(loc='upper right', bbox_to_anchor=(1.25, 1))
plt.show()
```

```python
# Definir variables de características y objetivo
features = ['WEAPON SOURCE COUNTRY', 'WEAPON DEPLOYMENT LOCATION', 'Data.Source', 'Location.Cordinates.Latitude',
            'Location.Cordinates.Longitude', 'Data.Magnitude.Body', 'Data.Magnitude.Surface', 'Location.Cordinates.Depth',
            'Data.Yeild.Lower', 'Data.Yeild.Upper', 'Date.Year']
target = 'Data.Yeild.Upper'
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(data[features], data[target], test_size=0.2, random_state=42)
```

```python
# Preprocesamiento e Ingeniería de Atributos
numeric_features = ['Location.Cordinates.Latitude', 'Location.Cordinates.Longitude', 'Data.Magnitude.Body',
                    'Data.Magnitude.Surface', 'Location.Cordinates.Depth', 'Date.Year']
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

categorical_features = ['WEAPON SOURCE COUNTRY', 'WEAPON DEPLOYMENT LOCATION', 'Data.Source']
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])
```

```python
import warnings

# Desactivar las advertencias de convergencia
warnings.filterwarnings("ignore", category=ConvergenceWarning)
```

```python
# Modelos a evaluar
models = [
    ('Ridge', Ridge()),
    ('Lasso', Lasso()),
    ('RandomForest', RandomForestRegressor())
]

best_model = None
best_mse = float('inf')

for name, model in models:
    pipeline = Pipeline(steps=[('preprocesador', preprocessor), ('modelo', model)])
    
    # Validación Cruzada
    scores = -cross_val_score(pipeline, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
    mean_mse = np.mean(scores)
    
    print(f"Modelo: {name}, MSE Promedio: {mean_mse:.2f}")
    
    # GridSearch para la optimización de hiperparámetros
    param_grid = {}
    if name in ['Ridge', 'Lasso']:
        param_grid['modelo__alpha'] = [0.01, 0.1, 1.0, 10.0]
    if name == 'RandomForest':
        param_grid['modelo__n_estimators'] = [50, 100, 200]
    
    grid_search = GridSearchCV(pipeline, param_grid, cv=3, scoring='neg_mean_squared_error')
    grid_search.fit(X_train, y_train)

    print("Mejor Modelo:", grid_search.best_estimator_)
    print("Mejor MSE:", -grid_search.best_score_)

    if -grid_search.best_score_ < best_mse:
        best_model = grid_search.best_estimator_
        best_mse = -grid_search.best_score_
```

```python
# Evaluar el mejor modelo en el conjunto de prueba
y_pred = best_model.predict(X_test)
test_mse = mean_squared_error(y_test, y_pred)
print("MSE en el conjunto de prueba del mejor modelo:", test_mse)
```
