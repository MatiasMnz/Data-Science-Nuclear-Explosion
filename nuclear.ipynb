```python
# Pregunta 1: Estacionalidad en la frecuencia de explosiones nucleares a lo largo de los aos
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='Date.Year')
plt.title('Estacionalidad en la Frecuencia de Explosiones Nucleares a lo largo de los Aos')
plt.xlabel('Ao')
plt.ylabel('Cantidad de Explosiones')
plt.xticks(rotation=45)
plt.show()
```

#### Breve Analisis
El grfico muestra la cantidad de explosiones nucleares registradas en cada ao. Se observa que la frecuencia de explosiones nucleares ha variado a lo largo de los aos. Algunos aos presentan picos de explosiones nucleares, mientras que en otros aos la cantidad es menor.

```python
# Insight 1: Estacionalidad en la Frecuencia de Explosiones Nucleares a lo largo de los Aos
print("Insight 1: Estacionalidad en la Frecuencia de Explosiones Nucleares a lo largo de los Aos")
year_with_max_explosions = data['Date.Year'].value_counts().idxmax()
max_explosions = data['Date.Year'].value_counts().max()
year_with_min_explosions = data['Date.Year'].value_counts().idxmin()
min_explosions = data['Date.Year'].value_counts().min()
print(f"Ao con mayor nmero de explosiones: {year_with_max_explosions} con {max_explosions} explosiones.")
print(f"Ao con menor nmero de explosiones: {year_with_min_explosions} con {min_explosions} explosiones.")
```

```python
# Pregunta 1: Se puede desarrollar un modelo preciso para predecir la potencia de una explosin nuclear en funcin de diversas variables?
print("Pregunta 1: Modelo de Prediccin de Potencia")
avg_potency = data['Data.Yeild.Upper'].mean()
max_potency = data['Data.Yeild.Upper'].max()
min_potency = data['Data.Yeild.Upper'].min()

plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='Data.Yeild.Upper', bins=20, kde=True)
plt.title('Distribucin de Potencia de Explosiones Nucleares')
plt.xlabel('Potencia (kilotones)')
plt.ylabel('Frecuencia')
plt.show()
```

#### Breve Anlisis

```python
# Insight
print("Insight 1: Modelo de Prediccin de Potencia")
print(f"Potencia promedio de explosiones nucleares: {avg_potency:.2f} kilotones")
print(f"Potencia mxima registrada: {max_potency:.2f} kilotones")
print(f"Potencia mnima registrada: {min_potency:.2f} kilotones")
```

```python
# Definir variables de caractersticas y objetivo
features = ['WEAPON SOURCE COUNTRY', 'WEAPON DEPLOYMENT LOCATION', 'Data.Source', 'Location.Cordinates.Latitude',
            'Location.Cordinates.Longitude', 'Data.Magnitude.Body', 'Data.Magnitude.Surface', 'Location.Cordinates.Depth',
            'Data.Yeild.Lower', 'Data.Yeild.Upper', 'Date.Year']
target = 'Data.Yeild.Upper'
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(data[features], data[target], test_size=0.2, random_state=42)
```

```python
# Preprocesamiento e Ingeniera de Atributos
numeric_features = ['Location.Cordinates.Latitude', 'Location.Cordinates.Longitude', 'Data.Magnitude.Body',
                    'Data.Magnitude.Surface', 'Location.Cordinates.Depth', 'Date.Year']
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

categorical_features = ['WEAPON SOURCE COUNTRY', 'WEAPON DEPLOYMENT LOCATION', 'Data.Source']
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])
```

```python
import warnings

# Desactivar las advertencias de convergencia
warnings.filterwarnings("ignore", category=ConvergenceWarning)
```

```python
# Modelos a evaluar
models = [
    ('Ridge', Ridge()),
    ('Lasso', Lasso()),
    ('RandomForest', RandomForestRegressor())
]

best_model = None
best_mse = float('inf')

for name, model in models:
    pipeline = Pipeline(steps=[('preprocesador', preprocessor), ('modelo', model)])
    
    # Validacin Cruzada
    scores = -cross_val_score(pipeline, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
    mean_mse = np.mean(scores)
    
    print(f"Modelo: {name}, MSE Promedio: {mean_mse:.2f}")
    
    # GridSearch para la optimizacin de hiperparmetros
    param_grid = {}
    if name in ['Ridge', 'Lasso']:
        param_grid['modelo__alpha'] = [0.01, 0.1, 1.0, 10.0]
    if name == 'RandomForest':
        param_grid['modelo__n_estimators'] = [50, 100, 200]
    
    grid_search = GridSearchCV(pipeline, param_grid, cv=3, scoring='neg_mean_squared_error')
    grid_search.fit(X_train, y_train)

    print("Mejor Modelo:", grid_search.best_estimator_)
    print("Mejor MSE:", -grid_search.best_score_)

    if -grid_search.best_score_ < best_mse:
        best_model = grid_search.best_estimator_
        best_mse = -grid_search.best_score_
```

```python
# Evaluar el mejor modelo en el conjunto de prueba
y_pred = best_model.predict(X_test)
test_mse = mean_squared_error(y_test, y_pred)
print("MSE en el conjunto de prueba del mejor modelo:", test_mse)
```
